```python
import bagpy
import numpy
import pandas
import allantools
import matplotlib.pyplot as plt
bag_read=bagpy.bagreader("C:\\Users\\nitin\\OneDrive\\Desktop\\RSN_DATA\\nitin.bag")
bag_string=bag_read.message_by_topic('/imu')
data_read=pandas.read_csv(bag_string)
```

```python
gyro_x=data_read["angular_velocity.x"]
gyro_y=data_read["angular_velocity.y"]
gyro_z=data_read["angular_velocity.z"]
data_time=data_read["Header.stamp.secs"]
time=data_time-data_time[0]
```

```python
fig, ax = plt.subplots()
plt.plot(time,gyro_x,label = "Gyro_x")
plt.plot(time,gyro_y,label = "Gyro_y")
plt.plot(time,gyro_z,label = "Gyro_z")
plt.xlabel('Time(sec)')
plt.ylabel('Gyro rad/s')
plt.ylim(-0.02,0.02)
plt.title("Gyro vs Time")
plt.legend()
plt.show()
```
