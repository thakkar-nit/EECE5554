```python
import bagpy
import numpy
import pandas
import allantools
import matplotlib.pyplot as plt
bag_read=bagpy.bagreader("C:\\Users\\nitin\\OneDrive\\Desktop\\RSN_DATA\\LocationB.bag")
bag_string=bag_read.message_by_topic('/vectornav')
data_read=pandas.read_csv(bag_string)
```

```python
data=data_read["data"]
data_time=data_read["header.stamp.secs"]
data_mani=numpy.array(data)
x_gyro=[]
y_gyro=[]
z_gyro=[]
time=[]
for splitting in data_mani:

    splitted=splitting.split("*")
    actual=splitted[0].split(",")
    # print(actual[10])
    x_gyro.append(float(actual[7]))
    y_gyro.append(float(actual[8]))
    z_gyro.append(float(actual[9]))

time=data_time-data_time[0]
```

```python
fig, ax = plt.subplots()
plt.plot(time,x_gyro,label = "Linear_acc_x")
plt.plot(time,y_gyro,label = "Linear_acc_y")
plt.plot(time,z_gyro,label = "Linear_acc_z")
plt.xlabel('Time(sec)')
plt.ylabel('Linear acceleration m*2/s')
plt.title("Linear Acceleration vs Time")
plt.legend()
plt.show()
```
