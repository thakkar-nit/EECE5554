```python
import bagpy
import numpy
import pandas
import allantools
import matplotlib.pyplot as plt
bag_read=bagpy.bagreader("C:\\Users\\nitin\\OneDrive\\Desktop\\RSN_DATA\\LocationB.bag")
bag_string=bag_read.message_by_topic('/vectornav')
data_read=pandas.read_csv(bag_string)
```

```python
data=data_read["data"]
data_time=data_read["header.stamp.secs"]
data_mani=numpy.array(data)
x_gyro=[]
y_gyro=[]
z_gyro=[]
time=[]
for splitting in data_mani:

    splitted=splitting.split("*")
    actual=splitted[0].split(",")
    # print(actual[10])
    x_gyro.append(float(actual[10]))
    y_gyro.append(float(actual[11]))
    z_gyro.append(float(actual[12]))

time=data_time-data_time[0]
```

```python
fig, ax = plt.subplots()
plt.plot(time,x_gyro,label = "Gyro_x")
plt.plot(time,y_gyro,label = "Gyro_y")
plt.plot(time,z_gyro,label = "Gyro_z")
plt.xlabel('Time(sec)')
plt.ylabel('Gyro rad/s')
plt.ylim(-0.02,0.02)
plt.title("Gyro vs Time")
plt.legend()
plt.show()
```
